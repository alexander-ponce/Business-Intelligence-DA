```python
# Function to calculate average age
def average_age(dataframe):
    return dataframe['age'].mean()
```

```python
# Function to calculate gender distribution
def gender_distribution(dataframe):
    male_count = dataframe[dataframe['sex'] == 'male'].shape[0]
    female_count = dataframe[dataframe['sex'] == 'female'].shape[0]
    return male_count, female_count
```

```python
# Function to calculate average yearly medical charges
def average_charges(dataframe):
    return dataframe['charges'].mean()
```

```python

# Function to list unique regions
def unique_regions(dataframe):
    return dataframe['region'].unique()
```

```python
# Perform the calculations
avg_age = average_age(df)
male_count, female_count = gender_distribution(df)
avg_charges = average_charges(df)
regions = unique_regions(df)
```
